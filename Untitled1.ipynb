```python
df=pd.read_csv(r'/content/survey.csv')
df.head()
```

```python
#Finding number of rows and attributes

print(df.shape[0])
print(df.shape[1])
```

There are 1259 rows and 27 attributes

```python
# Trying to find number of missing data:
print(round((df.isna().sum()/df.shape[0])*100,2))
```

```python
df.drop(['comments','state'],axis=1)
```

We observe that the comments attribute has many missing values , it is fair to say that it is optional and not important to our analysis.

```python

print(df['Age'].mean())
print(df['Age'][390])

sns.boxplot(df['Age'])
print((np.where(df['Age']>60)))
print((np.where(df['Age']<21)))
```

We observe there is alot of data in which Age>60 and Age<18 , these must be faulty data

```python
#Dropping faulty data
df.drop(df[(df['Age']>60) | (df['Age']<21)].index,inplace=True)

print(df['Age'].mean())


sns.boxplot(df['Age'])
print((np.where(df['Age']>60)))
print((np.where(df['Age']<21)))
```

We dropped extreme value and we can now observe we have a few outliers remaining.

```python
print(df['Age'].describe())
```

```python
#IQR = Q3-Q1
Q1=df['Age'].describe()[4]
Q3=df['Age'].describe()[6]
IQR=Q3-Q1

#print((df['Age']>(Q3+1.5*IQR)) | (df['Age']<(Q1-1.5*IQR)).sum())

df.drop(df[(df['Age']>(Q3+1.5*IQR)) | (df['Age']<(Q1-1.5*IQR))].index,inplace=True)
sns.boxplot(df['Age'])
```
